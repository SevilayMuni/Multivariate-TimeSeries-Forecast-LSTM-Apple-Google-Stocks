```python
import tensorflow as tf
import os
import numpy as np
import random

SEED = 0
```

```python
def set_seeds(seed = SEED):
    os.environ['PYTHONHASHSEED'] = str(seed)
    random.seed(seed)
    tf.random.set_seed(seed)
    np.random.seed(seed)
```

```python
def set_global_determinism(seed = SEED):
    set_seeds(seed = seed)

    os.environ['TF_DETERMINISTIC_OPS'] = '1'
    os.environ['TF_CUDNN_DETERMINISTIC'] = '1'
    
    tf.config.threading.set_inter_op_parallelism_threads(1)
    tf.config.threading.set_intra_op_parallelism_threads(1)

# Call the above function with seed value
set_global_determinism(seed = SEED)
```

## Important !

1. Please run cells above before executing any other code.
2. Model training might become slower since the code is deterministic -- a tradeoff.
3. The code belongs to stack overflow user Dan to [Question](https://stackoverflow.com/questions/36288235/how-to-get-stable-results-with-tensorflow-setting-random-seed)

### Information on Dataset:
Variables: 
- Open: the price when market opened in the morning.
- Close: the price when market closed in the afternoon.
- High: the highest price during that trading day.
- Low: the lowest price during that trading day.
- Volume: stock's number of shares traded that day. For instance, the stock trading volume would refer to the number of shares of security traded between its daily open and close. Trading volume, and changes to volume over the course of time, are important inputs for technical traders.
- Adj Close (Adjusted Close): price adjusted to make prices comparable over time. It is standard benchmark utilized by investors to track share's performance over time.

```python
# Determine end and start dates for dataset download
end = datetime.now()
start = datetime(end.year - 20, end.month, end.day)

# Download Apple's dataset btw 2004 - 2024
df_apple = yf.download('AAPL', start, end)
df_apple
```

```python
# Rename columns of Apple DF
column_dict = {'Open': 'open', 'High' : 'high', 'Low' : 'low',
               'Close' : 'close', 'Adj Close' : 'adj_close', 'Volume' : 'volume'}

df_apple = df_apple.rename(columns = column_dict)
df_apple.index.names = ['date']
df_apple.head(5)
```

<div style="background: linear-gradient(45deg, #567d9c, #90a3b4); padding: 15px; border-radius: 20px; text-align: center; background-color: black; color: black;"> <h1 style="font-size: 24px;"> Feature Engineering based on Domain Knowledge </h1>  
    
</div>

Various established technical indicators have been utilized by domain experts to analyze performance of shares. 
Garman-Klass Volatility, RSI, Bollinger Bands, ATR, MACD, Dollar Volume can be given as technical indicator examples. Since we want to forecast closing price of the share of interest, we will utilize only relevant technical indicators: Garman-Klass Volatility, Dollar Volume. In addition, we will calculate rolling sums.

1. Garman Klass is a volatility estimator that incorporates open, low, high, and close prices of a security. Therefore, our new feature will account for several variables. 
2. Dollar Volume is simply the number of shares traded multiplied by the price. It is an alternative version of volume designed to account for money flowing into and out of an asset on a particular day.
3. On-balance volume (OBV) is a simple accumulation-distribution tool that tallies up and down volume, creating a smooth indicator line to predict when major market moves might occur based on changes in relative trading volume. OBV works best when testing around major highs and lows to measure for possible breakouts and breakdowns. Investors can use OBV to provide many key predictions, such as a bullish divergence predicting the price will break resistance or a bearish divergence predicting a rally will stall or reverse.
3. Moving average convergence/divergence (MACD) is a technical indicator to help investors identify market entry points for buying or selling. The MACD line is calculated by subtracting the 26-period exponential moving average (EMA) from the 12-period EMA.
4. The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses.

```python
# Calculate Garman Klass Volatility
df_apple['garman_klass_volatility'] = ((np.log(df_apple['high']) - 
                                    np.log(df_apple['low']))**2)/2 -(
    2*np.log(2)-1)*((np.log(df_apple['adj_close'])-np.log(df_apple['open']))**2)

df_apple.head()
```

```python
# Create dollar volume column
df_apple['dollar_volume'] = (df_apple['adj_close'] * df_apple['volume'])/1e6

# Check whether column creations went smoothly
print(df_apple.columns)
```

```python
# Define function to calculate 'On Balance Volume (OBV)'
def On_Balance_Volume(Close, Volume):
    change = Close.diff()
    OBV = np.cumsum(np.where(change > 0, Volume, np.where(change < 0, -Volume, 0)))
    return OBV

# Create OBV column
obv = On_Balance_Volume(df_apple['close'], df_apple['volume'])
df_apple['obv'] = obv
```

```python
# Calculate the MACD

df_apple['macd'] = df_apple['close'].ewm(
    span = 12, adjust = False).mean() - df_apple['close'].ewm(
    span = 26, adjust = False).mean()
df_apple.head(5)
```

```python
# Create rolling sum for certain time periods(see ma_day list) 

ma_day = [3, 5, 10, 12, 20]

for ma in ma_day:
        column_name = f'ma_{ma}_days'
        df_apple[column_name] = df_apple['adj_close'].rolling(ma).mean()

# Check DF after feature engineering process
df_apple.tail()
```

```python
# Select sub-part of DF based on heatmap of correlation 
apple_features = ['adj_close', 'garman_klass_volatility', 'dollar_volume', 'obv', 'ma_3_days']
apple_dset = df_apple.filter(apple_features)
apple_dset.dropna(axis = 0, inplace = True)
print(apple_dset.shape)
```

```python
# Split train-test dsets wrt 80% - 20%
q_80 = int(len(apple_dset) * 0.8)

apple_train = apple_dset[:q_80]
apple_test = apple_dset[q_80:]

apple_train.shape, apple_test.shape
```

```python
# Scale both train and test dsets before training the model
scaler = MinMaxScaler(feature_range = (0, 1))
apple_train_scaled = scaler.fit_transform(apple_train)
apple_test_scaled = scaler.transform(apple_test)
apple_train_scaled
```

```python
# Reform datasets since it is time series data and will be fed into LSTM model
def createXY(dataset, n_past):
    dataX = []
    dataY = []
    for i in range(n_past, len(dataset)):
            dataX.append(dataset[i - n_past:i, 0:dataset.shape[1]])
            dataY.append(dataset[i,0])
    return np.array(dataX),np.array(dataY)

trainX, trainY = createXY(apple_train_scaled, 21) # model will be trained w/ 21 days time frames
testX, testY = createXY(apple_test_scaled, 21)
```

```python
# Check dataset shapes are compatible and consistent
trainX.shape, trainY.shape, testX.shape, testY.shape
```

<div style="background: linear-gradient(45deg, #567d9c, #90a3b4); padding: 15px; border-radius: 20px; text-align: center; background-color: black; color: black;"> <h1 style="font-size: 24px;"> Model Architecture and GridSearchCV </h1>  
    
</div>

## !!! GridSearchCV and model training won't be executed in this final notebook version since it takes very long time for my old laptop :(

```python
def build_model(optimizer):
    grid_model = Sequential()
    grid_model.add(LSTM(60, return_sequences = True, input_shape = (21, 5)))
    grid_model.add(LSTM(60))
    grid_model.add(Dropout(0.1))
    grid_model.add(Dense(1))
    grid_model.compile(loss = 'mse', optimizer = optimizer)
    return grid_model


grid_model = KerasRegressor(build_fn = build_model,
                            verbose = 1, validation_data = (testX, testY))
parameters = {'batch_size' : [20, 26],
              'epochs' : [15, 20],
              'optimizer' : ['adam', 'Adadelta']}
grid_search  = GridSearchCV(estimator = grid_model,
                            param_grid = parameters,
                            cv = 4)
```

```python
grid_search = grid_search.fit(trainX, trainY)
```

```python
grid_search.best_params_
```

### Note: The search yielded {'batch_size': 26, 'epochs': 20, 'optimizer': 'adam'} as best parameters.

```python
# Create trained model with best parameters yielded from GridSearchCV
apple_model = grid_search.best_estimator_.model
```

```python
# Save the model !!!
apple_model.save('Apple-LSTM-Model.h5')
print('Model is saved to disk :)')
```

```python
# Load previously saved model
apple_model = tf.keras.models.load_model('Apple-LSTM-Model.h5')
apple_model.summary()
```

```python
# Generate prediction and check its size
prediction = apple_model.predict(testX)

print('Prediction Shape:', prediction.shape)
```

```python
# Change data shape to use inverse_transform
prediction_array = np.repeat(prediction, 5, axis = -1)
original_array = np.repeat(testY, 5, axis = -1)
prediction_array
```

```python
# We just need the first column(adj_close) after inverse transform, so use [:,0] at the end
pred = scaler.inverse_transform(np.reshape(prediction_array,(len(prediction), 5)))[:,0]
original = scaler.inverse_transform(np.reshape(original_array,(len(testY), 5)))[:,0]

# Make sure dataset sizes are consistent
pred.shape, original.shape
```

## Information on evaluation metrics commonly applied for forecasting models:

MAE is defined as the average of absolute difference between forecasted values and true values.


The MAE does not give larger errors a higher weight, but, when used as a loss function in a machine learning model, it can cause convergence issues during the learning phase.

MAPE is defined as the percentage of the average of absolute difference between forecasted values and true values, divided by true value.

A MAPE under 10% shows high accuracy, 10-20% shows good forecast results, 20-50% shows reasonable forecast results, and greater than 50% shows incorrect results.


```python
# Calculate common evaluation metrics for forecasting model
mae_1 = mean_absolute_error(original, pred).round(4)
mape_1 = mean_absolute_percentage_error(original, pred).round(4)
R2_1 = r2_score(original, pred).round(4)

print('MAE =', mae_1, '\nMAPE =', mape_1, '\nR2 score =', R2_1)
```

```python
def build_2_model(optimizer):
    grid_model = Sequential()
    grid_model.add(LSTM(128, return_sequences = True, input_shape = (21, 5)))
    grid_model.add(LSTM(64))
    grid_model.add(Dense(10))
    grid_model.add(Dense(1))
    grid_model.compile(loss = 'mse', optimizer = optimizer)
    return grid_model


grid_model = KerasRegressor(build_fn = build_2_model, verbose = 1, validation_data = (testX, testY))
parameters = {'batch_size' : [16, 20, 26], 'epochs' : [10, 15, 20], 'optimizer' : ['adam', 'Adadelta']}
grid_search  = GridSearchCV(estimator = grid_model, param_grid = parameters, cv = 4)
```

```python
grid_search = grid_search.fit(trainX, trainY)
```

```python
grid_search.best_params_
```

### Note: The search yielded {'batch_size': 16, 'epochs': 20, 'optimizer': 'adam'} as best parameters.

```python
# Save the model !!!
apple_model2 = grid_search.best_estimator_.model
apple_model2.save('Apple-2nd-LSTM-Model.h5')
print('Model is saved to disk :)')
```
